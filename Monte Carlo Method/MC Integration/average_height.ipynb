## Human Average Height Example ##
A pracitcal example to understand the fundamental idea behind the Monte Carlo Method.

Suppose we want to compute the average height of *100 million* humans on earth, instead of measuring each one we randomply select a **fraction** of them, and we apply the Monte Carlo idea by: 

 **1- Random Sampling** : Randomply select a sub-group of people.\
 **2- Modelling** : Compute the Arithmetical average on the sample.\
 **3- Simulation**: Repeat step 1 & 2.\
 **4- Statistical** Analysis : Display results in a histogram & and compute the overall average.

```python
# Importing the module and generating our data

import numpy as np

height_data = np.random.normal(170, 20, 10_000_000)
sub_height_data = np.random.choice(height_data, 1_000_000)
```

```python
list_heights = []

for i in range(1, 10_000): # Step 3: Simulating
    sample = np.random.choice(sub_height_data, 100_000) # Step 1: Random Sampling
    average_height_sample = np.average(sample)  # Step 2: Modelling

    list_heights.append(average_height_sample)

average_height_mt = np.average(list_heights)

print(f"Arithmetical Avergage: {np.average(height_data):.4f} [cm]")
print(f"Random Sampling Avergage: {average_height_mt:.4f} [cm]")
print(f"Error %: {100*(np.average(height_data)-average_height_mt)/np.average(height_data):.4f} [%]")
```

```python
import matplotlib.pyplot as plt
plt.hist(list_heights, bins=25, color ='green', alpha = 0.7, edgecolor='black')
plt.show()
```
